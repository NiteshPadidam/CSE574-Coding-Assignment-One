```python
Y = df_p["sex"]
X = df_p[['species','island','bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']]
X.head()
```

```python
loss = []
class LogitRegression ():

    def __init__ (self,learning_rate,iterations):
        self.learning_rate = learning_rate
        self.iterations =iterations


    def fit (self,X,Y):
        self.X = X.values
        print(self.X.shape)
        self.Y = Y.values
        self.wts = np.random.uniform(0, 1,[X.shape[1],1])
        print('initial wts',self.wts)
        for iter in range(self.iterations):
            pred = self.sigmoid(np.dot((self.X),(self.wts)))
            K = self.cost(Y.values,pred)
            loss.append(K)
            self.wts = self.wts - (self.gradient_descent(self.X,self.Y,pred))
        print('final wts',self.wts)

    def sigmoid (self,z):
        value = 1 / (1 + np.exp(-z))
        return value

    def cost (self,Y,pred):
        m = len(pred)
        K = -(np.sum(Y * np.log(pred) + (1 - Y) * np.log(1 - pred)))
        return K

    def gradient_descent (self,X,Y,pred):
        m = len(pred)
        descent = np.dot(X.T, (pred - Y))
        return self.learning_rate * descent

    def predict(self , X):
        print(self.wts)
        print(self.learning_rate)
        print(self.iterations)
        prob = self.sigmoid(np.dot(X.values, self.wts))
        preds = [1 if p>=0.5 else 0 for p in prob]
        return preds
```

```python
X_train, X_test,Y_train, Y_test = split_data_train_test(X,Y, 0.8,69)
print(set(Y_test['sex'].to_list()))
X_train.shape
# Y_test
```

```python
loss = []
model = LogitRegression(learning_rate = 1e-6,iterations= 100000)# 0.746268656716418
# type(X_train.values)
model.fit(X_train, Y_train)
len(loss)
```

```python
predicted = model.predict(X_test)
accuracy = np.mean(predicted == Y_test['sex'].values)
predicted == Y_test['sex'].to_list()
accuracy
```
